```python
import pandas as pd
from prophet import Prophet 
from matplotlib import pyplot as plt
plt.figure(figsize=(10,10))
```

```python
data = pd.read_feather('tavbase/Sample.feather')
data
```

```python
## Aula 10 - RegressÃ£o (com sazonalidades)
regressao_segment = pd.DataFrame()
print(data)
for segment in data['Segment'].unique():
    print(segment)
    regressao = data[data['Segment'] == segment][
        ['Order Date Month', 'Sales']
    ].groupby('Order Date Month')['Sales'].sum().reset_index()

    regressao = regressao.rename(
        columns={
            'Order Date Month': 'ds', 'Sales': 'y'
        }
    )
    
    m = Prophet().fit(regressao)
    future = m.make_future_dataframe(periods=12, freq='MS')
    forecast = m.predict(future)

    forecast['Segment'] = segment
    fig1 = m.plot(forecast)
    plt.show()

    regressao_segment = regressao_segment.append(
        forecast[['Segment','ds','yhat','yhat_lower','yhat_upper']],
        ignore_index = True
    )
regressao_segment    
```

```python
regressao_segment.to_feather('tavbase/regressao_segment.feather')
```

```python
regressao_region = pd.DataFrame()
for region in data['Region'].unique():
    print(region)
    regressao = data[data['Region'] == region][
        ['Order Date Month', 'Sales']
    ].groupby('Order Date Month')['Sales'].sum().reset_index()
    regressao = regressao.rename(columns={
        'Order Date Month': 'ds', 'Sales': 'y'
    })

    m = Prophet().fit(regressao)
    future = m.make_future_dataframe(periods=12, freq='MS')
    forecast = m.predict(future)

    forecast['Region'] = region
    fig1 = m.plot(forecast)
    plt.show()
    
    regressao_region = regressao_region.append(
        forecast[['Region','ds','yhat','yhat_lower','yhat_upper']],
        ignore_index = True
    )
regressao_region    
```

```python
regressao_region.to_feather('tavbase/regressao_region.feather')
```
